### 0. Import libraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

sns.set()
```

### 1. Analyze the LZ77 and LZW algorithms for each type of text

##### 1.1. lorem ipsum texts

```python
measurents_lorem_ipsum = pd.read_csv("./data/measurents_lorem_ipsum_texts.csv", index_col=0)
measurents_lorem_ipsum
```

```python
measurents_lorem_ipsum.describe()
```

##### 1.2. english literature texts

```python
measurements_english_literature = pd.read_csv("./data/measurents_english_literature_texts.csv", index_col=0)
measurements_english_literature
```

```python
measurements_english_literature.describe()
```

##### 1.3. programming source code

```python
measurents_programming_source_code = pd.read_csv("./data/measurents_programming_source_code.csv", index_col=0)
measurents_programming_source_code
```

```python
measurents_programming_source_code.describe()
```

### 2. Compute correlation between all statistics

```python
all_measurements = pd.concat([measurents_lorem_ipsum, measurements_english_literature, measurents_programming_source_code])
all_measurements.corr()
```

```python
sns.heatmap(all_measurements.corr(), annot=True, cmap="viridis_r")
plt.title("Pearson correlation between all measurements")
```

### 3. Analyze the compression rates

```python
all_measurements["type of text"] = 10 * ["Lorem Ipsum"] + 10 * ["English literature"] + 10 * ["Python code"]

sns.boxplot(data=all_measurements, y="type of text", x="LZ77 compression rates", hue="type of text", palette="viridis_r")
plt.xlabel("LZ77 compression rates (%)")
```

```python
all_measurements["type of text"] = 10 * ["Lorem Ipsum"] + 10 * ["English literature"] + 10 * ["Python code"]

sns.boxplot(data=all_measurements, y="type of text", x="LZW compression rates", hue="type of text", palette="viridis_r")
plt.xlabel("LZW compression rates (%)")
```
